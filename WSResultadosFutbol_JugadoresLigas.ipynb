En este proyecto trabajaré con las librerias de python: requests, bs4, pandas, sqlachemy y psycopg2

Estos datos formarán parte de la tabla 'jugadores' en la base de datos análisis futbol en postgresql, donde hare, entre otros, el modelado de datos junto a las tablas  y 'equipos' y regiones

```python
from bs4 import BeautifulSoup
```

```python
# esta funcion es para obtener los links de cada una de las ligas de futbol de interés
def obtener_websites_equipos(link_ppal):
    re = requests.get(link_ppal)
    sopa = BeautifulSoup(re.text, 'lxml')
    sopa_equipos_liga = sopa.find('table', attrs= {'id':'tabla2'}).find_all('tr')
    sopa_equipos_liga.pop(0)
    links1 = [url.a.get('href') for url in sopa_equipos_liga]
    links2 = ["https://www.resultados-futbol.com"+url for url in links1]
    return links2
```

```python
# función para extraer los datos de cada jugador de todos los equipos de la liga dada
def armar_plantilla(link):  
    
    
    r = requests.get(link)
    sopa = BeautifulSoup(r.text, 'lxml')
    ptll_soup = sopa.find_all('tr', attrs= {'itemprop':'employee'})
    
        
    plantilla = []
    i = 0
    
    while i <  len(ptll_soup):
                    
                   
        print(f"estoy extrayendo los datos del jugador {i}")
        
                #equipo
        equipo =  str(link[34:])
                #posicion
        posicion = ptll_soup[i].find_previous_sibling('tr', attrs = {'class':'first-child'}).get_text(strip = True)
                #nombre
        nombre = ptll_soup[i].find('span').text
               #numero
        numero_jugador = ptll_soup[i].find('td', attrs= {"class":"num"}).text
                #edad
        edad_jugador = ptll_soup[i].find('td', attrs={'itemprop':"birthDate"}).text
                #nacionalidad
        tag_nacionalidad = ptll_soup[i].find('td', attrs= {'itemprop':'nationality'})
        str_nacionalidad = str(tag_nacionalidad)
        nacionalidad = str_nacionalidad[104:106]
                #Altura y peso
        altura_peso = ptll_soup[i].find_all('td', attrs= {'class':'dat'}, limit=2)
        altura= altura_peso[0].text
        peso= altura_peso[1].text

                #agregar nuevo elemento a la lista
        plantilla.extend([equipo])
        plantilla.extend([numero_jugador])
        plantilla.extend([posicion])
        plantilla.extend([nombre])
        plantilla.extend([edad_jugador])
        plantilla.extend([nacionalidad])
        plantilla.extend([altura])
        plantilla.extend([peso])

        i = i+1
        
    else:
        
        return plantilla
```

```python
liga = str(url[34:])
liga
```

```python
#aplico función al link de la liga para obtener los links de cada equipo que la conforma
links = obtener_websites_equipos(url)
```

```python
links
```

```python
#lista donde agrego los datos de cada jugador obtenidos con las función "armar_plantilla"
lis = []
for i in links:
    lis.append(armar_plantilla(i))
    
```

```python
len(lis)
```

```python
#hago una (1) lista con todos los datos
lista = []
i = 0
while i < len(lis):
    l1= lis[i]
    lista.extend(l1)
    i = i +1
    
```

```python
len(lista)
```

```python
lista
```

```python
# separo la lista en sublistas por jugador
n=8

plantilla_final =[lista[i:i + n] for i in range(0, len(lista), n)]
plantilla_final
```

```python
import pandas as pd
```

```python
#Expreso como DataFrame
plantillas_df = pd.DataFrame(plantilla_final)
plantillas_df
```

```python
#Nombro columnas
plantillas_df.columns=['equipo','numero','posicion', 'nombre','edad','nacionalidad','altura','peso']
```

```python
#establezco índice para evitar conflictos con la base de datos de destino
plantillas_df.set_index('equipo', inplace=True)
```

```python
# asigno una columna con el nombre de la liga a la cual pertenece cada jugador
plantillas_df= plantillas_df.assign(Liga= liga)
```

```python
plantillas_df.info()
```

```python
plantillas_df
```

```python
#cambio tipo de datos
plantillas_df['altura'] = pd.to_numeric(plantillas_df['altura'], errors= 'coerce')
plantillas_df['peso'] = pd.to_numeric(plantillas_df['peso'], errors= 'coerce')
plantillas_df['edad'] = pd.to_numeric(plantillas_df['edad'], errors= 'coerce')
```
